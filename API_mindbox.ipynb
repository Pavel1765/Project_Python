```python
import pandas as pd
import requests
import time
```

```python
API_URL = "https://api.mindbox.ru/v3/operations/sync?endpointId=#####################"
API_KEY = "###################"
def get_task_id():
   headers = {
       "Content-Type": "application/json; charset=utf-8",
       "Accept": "application/json",
       "Authorization": f"SecretKey {API_KEY}"
   }

   response = requests.post(API_URL, headers=headers)
   if response.status_code == 200:
       data = response.json()
       exportId = data.get('exportId')
       return exportId
   else:
       print(f"Ошибка: {response.status_code} - {response.text}")
       return None
    
def get_task_file(task_id):
   headers = {
       "Content-Type": "application/json; charset=utf-8",
       "Accept": "application/json",
       "Authorization": f"SecretKey {API_KEY}"
   }
   json_data = {"exportId": task_id}
    
   response = requests.post(API_URL, headers=headers, json= json_data)
   if response.status_code == 200:
       data = response.json()
       data_json= data.get('exportResult')
       # print(data_json['urls'][0])
       return data_json['urls'][0]
   else:
       print(f"Ошибка: {response.status_code} - {response.text}")
       return None
    
if __name__ == "__main__":
   task_id = get_task_id()
   while True:  # Бесконечный цикл
       time.sleep(50)  # Первоначальная пауза в 50 секунд
       try:
           task_file = get_task_file(task_id)
           print(f"ID задачи: {task_id}")
           print(f"Файл задачи: {task_file}")
           break  # Выход из цикла, если код выполнен успешно
       except KeyError as e:
           print(f"Ошибка: {e}. Ожидание дополнительных 50 секунд...")
           # time.sleep(50)  # Дополнительная пауза в 50 секунд
           # Цикл продолжится, и попытка будет повторена
```
